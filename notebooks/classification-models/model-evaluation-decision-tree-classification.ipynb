# Model evaluation of decision tree classification

## Importing the libraries

```python
import numpy as np
import pandas as pd
import os.path
```

## Load the dataset

```python
raw_data_file_path = '/Users/vishalsaxena/Documents/DS_Workspace/udemy_ml_algo_python/data/raw/'
dataset = pd.read_csv(os.path.join(raw_data_file_path, 'breast-cancer.csv'))
dataset.drop(columns=['Sample code number'], inplace=True)
dataset
```

```python
X = dataset.iloc[:,:-1].values
X
```

```python
y = dataset.iloc[:,-1].values
```

## Splitting the dataset into Train/Test

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
X_train, X_test
```

```python
y_test, y_train
```

## Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
```

```python
X_test = sc.transform(X_test)
```

## Training the model

```python
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier(random_state=0, criterion='entropy')
classifier.fit(X_train, y_train)
```

## Predicting the result set

```python
y_pred = classifier.predict(X_test)
```

```python
np.concatenate((y_pred.reshape(len(y_pred), 1), y_test.reshape(len(y_test), 1)), axis=1)
```

## Calculating the confusion matrix and accuracy metrics

```python
from sklearn.metrics import confusion_matrix, accuracy_score
confusion_matrix(y_test, y_pred)
```

```python
np.round(accuracy_score(y_test, y_pred), 2)
```
