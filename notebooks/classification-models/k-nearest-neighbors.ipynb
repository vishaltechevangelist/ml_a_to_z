# K-NN (K-nearest-neighbors)

## Importing libraries

```python
import numpy as np
import pandas as pd
import os.path
import matplotlib.pyplot as plt
```

## Load the dataset

```python
raw_data_file_path = '/Users/vishalsaxena/Documents/DS_Workspace/udemy_ml_algo_python/data/raw/'
dataset = pd.read_csv(os.path.join(raw_data_file_path, 'Social_Network_Ads.csv'))
dataset
```

```python
X = dataset.iloc[:,:-1].values
X
```

```python
y = dataset.iloc[:,-1].values
y
```

## Splitting the dataset into Train/Test

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
X_train, X_test
```

```python
y_train, y_test
```

## Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train_sc = sc.fit_transform(X_train)
X_train_sc
```

```python
X_test_sc = sc.transform(X_test)
X_test_sc
```

## Training the K-NN model with Training dataset 

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors=5, metric='minkowski', p=2)
classifier.fit(X_train_sc, y_train)
```

## Predicting new result with random test data

```python
classifier.predict(sc.transform([[30, 80000]]))
```

## Predicting the test set results

```python
y_pred = classifier.predict(X_test_sc)
y_pred
```

## Model evaluation

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

```python
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator=classifier, X=X_train_sc, y=y_train, cv=10)
accuracies.mean().round(2), accuracies.std().round(2)
```
