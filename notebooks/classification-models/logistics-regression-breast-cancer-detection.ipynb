# Breast Cancer detection using logistic regression

## Importing the libraries

```python
import numpy as np
import pandas as pd
from ucimlrepo import fetch_ucirepo
```

## Fetch breast cancer dataset

```python
breast_cancer_json = fetch_ucirepo(id=15)
breast_cancer_df = breast_cancer_json.data.features
breast_cancer_df
```

```python
breast_cancer_df['class'] = breast_cancer_json.data.targets
breast_cancer_df
```

## Handle the missing values

```python
breast_cancer_df.isna().sum()
```

## Drop missing values

```python
breast_cancer_df = breast_cancer_df.dropna()
breast_cancer_df.isna().sum()
```

```python
X = breast_cancer_df.iloc[:,:-1].values
X
```

```python
y = breast_cancer_df.iloc[:,-1].values
y
```

## Splitting the dataset into the Training/Test 

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
X_train, X_test
```

```python
y_train, y_test
```

## Training the logistics regression

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier.fit(X_train, y_train)
```

## Prediction the Test set results

```python
y_pred = classifier.predict(X_test)
y_pred
```

## Making the confusion matrix

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)
```

```python
np.equal(y_pred, y_test)
```

## Calculating accuracy score

```python
from sklearn.metrics import accuracy_score
np.round(accuracy_score(y_test, y_pred), 2)
```

## Computing the accuracy with k-fold cross validation

```python
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator = classifier, X = X_train, y = y_train, cv = 10)
accuracies.mean().round(2)*100, accuracies.std().round(2)*100
```
