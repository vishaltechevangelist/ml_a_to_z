# CatBoost

## Importing the library

```python
import numpy as np
import pandas as pd
import os.path
```

## Load the dataset

```python
raw_data_file_path = '/Users/vishalsaxena/Documents/DS_Workspace/udemy_ml_algo_python/data/raw/'
dataset = pd.read_csv(os.path.join(raw_data_file_path, 'breast-cancer.csv'))
dataset
```

## Drop the irrelevant column

```python
dataset = dataset.drop(columns=['Sample code number'])
dataset
```

## Segregate independent and dependent variable

```python
X = dataset.iloc[:,:-1].values
X
```

```python
y = dataset.iloc[:,-1].values
y
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
y = encoder.fit_transform(y)
y
```

## Splitting the dataset in Train/Test split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
X_train, X_test
```

```python
y_train, y_test
```

## Training XGBoost on the training set

```python
from catboost import CatBoostClassifier
classifier = CatBoostClassifier()
classifier.fit(X_train, y_train)
```

## Predicting the result set

```python
y_pred = classifier.predict(X_test)
y_pred
```

```python
np.concatenate((y_test.reshape(len(y_pred), 1), y_pred.reshape(len(y_test), 1)), axis=1)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

## K-fold validation

```python
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator=classifier, X=X_train, y=y_train, cv=10)
accuracies
```

```python
accuracies.mean(), accuracies.std().round(2)
```
